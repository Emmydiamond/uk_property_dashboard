```python
import pandas as pd
import os
import glob
```

```python
# Folder containing all yearly CSV files
folder_path = r"C:\Users\daret\Desktop\pricepaid"

# Get all csv files
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

print("Files found:", len(all_files))
print(all_files[:5])

# Read and combine
df_list = []

for file in all_files:
    temp_df = pd.read_csv(
        file,
        header=None,
        low_memory=False
    )

    df_list.append(temp_df)

df = pd.concat(df_list, ignore_index=True)

print("Combined Shape:")
print(df.shape)
```

```python
# Assign the correct column names manually

df.columns = [
    'Price',
    'Date of Transfer',
    'Postcode',
    'Property Type',
    'new_build_flag',
    'tenure',
    'Town/City',
    'District',
    'County',
    'extra1',
    'extra2'
]

# Remove unnecessary empty columns
df = df.drop(columns=['extra1','extra2'])

# Check result
df.head()
```

```python
# Remove rows where Price column literally contains "Price"

df = df[df['Price'] != 'Price']

# Reset index
df = df.reset_index(drop=True)

# Check first rows
df.head()
```

```python
import pandas as pd

# Remove repeated header rows
df = df[df['Price'] != 'Price']

# Reset index
df = df.reset_index(drop=True)

# Convert Price to numeric
df['Price'] = pd.to_numeric(df['Price'])

# Convert date column
df['Date of Transfer'] = pd.to_datetime(
    df['Date of Transfer'],
    dayfirst=True
)

# Create year/month features
df['year'] = df['Date of Transfer'].dt.year
df['month_num'] = df['Date of Transfer'].dt.month
df['month'] = df['Date of Transfer'].dt.month_name()

# Convert property type codes
property_map = {
    'D':'Detached',
    'S':'Semi-Detached',
    'T':'Terraced',
    'F':'Flat',
    'O':'Other'
}

df['property_type'] = df['Property Type'].map(property_map)

# Convert tenure codes
tenure_map = {
    'F':'Freehold',
    'L':'Leasehold'
}

df['tenure_type'] = df['tenure'].map(tenure_map)

# Convert new build codes
new_build_map = {
    'Y':'New',
    'N':'Existing'
}

df['new_build_status'] = df['new_build_flag'].map(new_build_map)

# Final dashboard dataset
dashboard_df = df[
[
'Price',
'Date of Transfer',
'year',
'month_num',
'month',
'property_type',
'new_build_status',
'tenure_type',
'Postcode',
'Town/City',
'District',
'County'
]
]

# Save prepared dashboard dataset
dashboard_df.to_csv(
    "uk_dashboard_data.csv",
    index=False
)

print("Dashboard dataset created successfully")
print(dashboard_df.info())

dashboard_df.head()
```
